```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def setDF(path_to_cpu, path_to_throughput):
    cpu = pd.read_csv(path_to_cpu)
    thr = pd.read_csv(path_to_throughput)
    fullDF = pd.concat([cpu.iloc[:, :-2], thr.iloc[:,1:]], axis=1)
    return fullDF
    
```
